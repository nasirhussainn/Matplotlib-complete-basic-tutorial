**importing matplotlib and then using inline magic**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**using simple data to plot x and y**

*color=black and makrer=diamond and linestyle=dotted*

the above shorthand is 'bD+'. This shorthand mean color linestyle and marker which is now + symbol


**Alpha** is used to set the transparency of line ranging between 0 and 1 at 0 line will be invisible

```python
x=[1,2,3,4,5]
y=[12,34,11,34,35]
plt.plot(x,y, 'bD', linestyle='dotted', alpha=1)
plt.xlabel("WeekDays")
plt.ylabel("Temperature")
plt.title("Weather")
```

***Working with Multiple Data***

```python
days=[1,2,3,4,5]
min_temp=[20,19,22,21,23]
max_temp=[30,35,31,40,33]
avg_temp=[25,26,27,30,27]
```

Plotting each data of min,max and avg temp with data on x-axi

```python
plt.plot(days,min_temp, label="Min")
plt.plot(days,max_temp, label="Max")
plt.plot(days,avg_temp, label="Avg")
```

Above Graph is not making any sense so make it more understandable by adding

-axis labels

-title

-grid

-label color 

```python
plt.plot(days,min_temp, label="Min", marker="D", linestyle="dotted")
plt.plot(days,max_temp, label="Max", marker="+", linestyle="dotted")
plt.plot(days,avg_temp, label="Avg", marker="*", linestyle="dotted")

plt.xlabel("WeekDays")
plt.ylabel("Temperature")
plt.title("Weather")

plt.legend(loc="best")
plt.grid()
```

***BAR CHART***

In below chart we are using profit and revenue of four US companies and plot it on bar chart. /

```python
companies=["OpenAI","MSFT","GOOGL","FB"]
revenue=[42,20,33,10]
profit=[32,15,15,5]
```

These are just dummy values now let make a bar graph

```python
plt.bar(companies,revenue)
plt.bar(companies,profit)
```

Now we can see that it is not that simple. So we have to make the companies in numeric to plot it on bar graph. Let see how we can do it.

```python
import numpy as np
```

```python
xpos=np.arange(len(companies))
xpos
```

The above numpy array has now list of companies as numeric form

```python
plt.bar(xpos,revenue)
```

But we do not want the number to appear on bar graph. So we replace the number with name of companies. 

```python
plt.xticks(xpos,companies)
plt.bar(xpos,revenue)
```

The reason why we do it becuase now we want to both revenue and profit side-by-side so we can move the bar easily becuase originally it is in numeric.

```python
plt.xticks(xpos,companies)
plt.bar(xpos,revenue)
plt.bar(xpos,profit)
```

`Now we will adjust this`

```python
plt.xticks(xpos,companies)
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4, label="Profit")

plt.xlabel("US Tech Comapnies")
plt.ylabel("Profie&Revenue (in bln)")
plt.title("US TECH GIANTS")

plt.legend()
plt.grid(color='r', linestyle=':', linewidth=0.5)
```

***HISTOGRAM***

We are using the following data to analyze the blood sugar of male and female patients

**For Blood Sugar Analysis**

*80-100 = Normal*

100-125 = Pre-Diabetic

125 and above = Diabetic

Now let use data to analyze

```python
blood_sugar_male=[45,100,120,100,128,80,135,70.111,110,84,144,99,130,92,89,119,129,61,101,122]
blood_sugar_female=[55,90,110,210,128,90,125,125,101,130,74,115,109,160,102,99,109,139,51,111,112]
```

Now just make a simple one using only entity

```python
plt.hist(blood_sugar_male)
```

bins is an important property which is used to set the range. Now let set it explicitly

```python
plt.hist(blood_sugar_male, bins=3, rwidth=0.9)
```

Now we have an idea about bins and rwidth now we properly analyze the blood sugar of male and female. That how many of them are normal, diabetic and pre-diabetic.

```python
plt.hist([blood_sugar_male,blood_sugar_female],bins=[80,100,120,150], rwidth=0.9, label=['male','female'], 
         color=['blue','purple'])
plt.legend(shadow=True)
plt.grid(color='r', linestyle=':', linewidth=0.5)
```

Now we can say that by seeing the histogram that 8 female patient have pre-diabetic and so on. You can use orientation=`horizontal` to convert the above in horizontal

***PIE CHART***

*Suppose we are tracking the sales of different flavors of ice cream at a particular ice cream shop for the month of July. Here are the sales figures:*

Vanilla: 300

Chocolate: 250

Strawberry: 150

Mint Chocolate Chip: 100

Cookies and Cream: 100

Butter Pecan: 50

Other: 50

```python
IceCream = ['Vanilla', 'Chocolate', 'Strawberry', 'Mint Chocolate Chip', 'Cookies and Cream', 'Butter Pecan', 'Other']
sales = [300, 250, 150, 100, 100, 50, 50]
```

plt.axis to make it look good and plt.show() to show only chart and hide other extra detail

```python
plt.axis("equal")
plt.pie(sales,labels=IceCream, radius=1.2, autopct='%0.0f%%')
plt.show()
```

Radius set the size of Pie chart

%0.0f% set the percentage. value after 0. show the number of decimal number that will be show on chart
in above we have 0.0 so there is 0 decimal number but if we set 0.3 then it show upt 3 decimal place

**explode** use to focus on specific pie

```python
plt.axis("equal")
plt.pie(sales,labels=IceCream, radius=1.2, autopct='%0.0f%%', explode=(0,0,0.2,0,0,0,0), startangle=145)
plt.show()
```
